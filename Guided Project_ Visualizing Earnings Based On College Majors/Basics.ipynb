```python
import pandas as pd 
import matplotlib.pyplot as plt 
%matplotlib inline

recent_grads = pd.read_csv("recent-grads.csv")
print(recent_grads.iloc[0])
print(recent_grads.head(5))
print(recent_grads.tail(2))
print(recent_grads.describe())
raw_data_count = 
recent_grads = recent_grads.dropna()
```

```python
recent_grads.plot(x='Sample_size', y='Median', kind='scatter' ,title='samplesize vs median' )
recent_grads.plot(x='Sample_size', y='Unemployment_rate', kind='scatter', title='Employed vs. Sample_size', figsize=(5,10))
ax = recent_grads.plot(x='Full_time', y='Median', kind='scatter')
recent_grads.plot(x='ShareWomen', y='Unemployment_rate', kind='scatter',title = 'sharewomen vs unemployement rate')
recent_grads.plot(x='Women', y='Median', kind='scatter', title = 'women vs median ')

recent_grads.plot(x='Men', y='Median', kind='scatter' , title = 'men vs median ')

ax.set_title('full_time vs. median')
```

```python
col = ["Sample_size", "Median", "Employed", "Full_time", "ShareWomen", "Unemployment_rate", "Men", "Women"]
fig = plt.figure(figsize = (10,20))
for i in range(1,5):
    ax = fig.add_subplot(4,1,i) 
    ax=recent_grads[col[i]].plot(kind='hist' , rot = '40' , title = col[i])

```
